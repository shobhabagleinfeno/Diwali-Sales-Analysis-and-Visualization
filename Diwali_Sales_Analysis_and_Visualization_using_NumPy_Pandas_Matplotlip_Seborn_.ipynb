## **Explorator Data Analysis**

### **Gender**

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df, hue="Gender", palette=['pink', 'blue'],label=False,dodge=False)
for bars in ax.containers:
  ax.bar_label(bars)

plt.title("Gender of the Customer")
plt.xlabel("Gender")
plt.ylabel("Count of the Gender")
plt.show()
```

***Fig 1. CountPlot of Customers based on there Gender***

*In the above graph it is clearly visible that the Females Customersare more in number than the Males  Customers*

```python
df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
gender_sales=df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=gender_sales,hue="Gender",palette=["pink","blue"],label=False,dodge=False,alpha=0.6)
plt.title("Gender wise Amount Spend")
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.show()
```

***Fig 2. Bar Plot of Amount spend by Customers based on there Gender***

*From the above chart it is clear that Female customers spent more than male customers*

## **Age**

```python
ax = sns.countplot(x='Age Group', data=df, hue="Gender",palette=['pink', 'blue'])
plt.title("Count of Age Group of the Customers by Gender")
plt.xlabel("Age Group")
plt.ylabel("Count of the Age Group")
plt.show()
```

***Fig 3. Shows the max number of age group of customers in our case the age group 26-35 is the max age group of customers come for buying***

```python
ax = sns.countplot(data=df, x="Age Group", hue="Gender", palette=["pink","blue"])

for bars in ax.containers:
  ax.bar_label(bars)
plt.title("Female and Male Customers Count based on there Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count of the Age Group")
plt.show()
```

***Fig 4. Shows the Count of Female and Male customers of a specific age group***

```python
age_group_palette = ['pink', 'blue', 'green', 'orange', 'purple', 'red', 'brown']
ax = sns.countplot(data=df, x="Age Group", hue="Age Group", palette=age_group_palette)

for bars in ax.containers:
  ax.bar_label(bars)

plt.title("Count of Age Group of the Customers")
plt.xlabel("Age Group")
plt.ylabel("Count of the Age Group")
plt.show()
```

***Fig 5. Shows the count of number of customers from that age group***

```python
age_group_palette = ['black', 'blue', 'green', 'orange', 'purple', 'red', 'brown']
age_sales=df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=age_sales,palette=age_group_palette)
plt.title("Age Group wise Amount Spend")
plt.xlabel("Age Group")
plt.ylabel("Amount")
plt.show()
```

***Fig 6. From the above chart it is clear that the age group 26-35 spend more amount than any other age group***

## **State**

```python
state_sales=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
sns.barplot(data=state_sales,x="State",y="Orders",palette="plasma")
plt.xticks(rotation=90)

plt.title("State wise Orders")
plt.xlabel("State")
plt.ylabel("Orders")
plt.show()
```

***Fig 7. Shows State wise Orders received***

```python
state_sales=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data=state_sales,x="State",y="Amount",palette="rocket")
plt.xticks(rotation=90)

plt.title("State wise Amount Spend")
plt.xlabel("State")
plt.ylabel("Amount")
plt.show()
```

***Fig 8 . Shows State wise Amount Spend***

## **Marital Status**

```python
ax = sns.countplot(data=df, x="Marital_Status", palette=["red","orange"])

for bars in ax.containers:
  ax.bar_label(bars)

plt.title("Count of Marital Status of the Customers")
plt.xlabel("Marital Status")
plt.ylabel("Count of the Marital Status")
plt.show()
```

***Fig 9.Shows the count of Married and Unmarried Customers***

***Fig 10. Shows Amount spend by Married and Unmarried***  


```python
marital_Gender_sales=df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data=marital_Gender_sales,x="Marital_Status",y="Amount",palette="icefire",hue="Gender")
plt.xticks(rotation=90)

plt.title("Marital Status wise Amount Spend by Gender")
plt.xlabel("Marital Status")
plt.ylabel("Amount")
plt.show()
```

***Fig 11. Shows Amount spend by Married ( men and women ) and Unmarried ( men and women )***

## **Product Category**

```python
ax=sns.countplot(data=df,x="Product_Category",palette="mako")

for bars in ax.containers:
  ax.bar_label(bars)

plt.xticks(rotation=90)
plt.title("Count of Product Category of the Customers")
plt.xlabel("Product Category")
plt.ylabel("Count of the Product Category")
plt.show()
```

***Fig 15. Shows the Amount spend on a Product Category***
